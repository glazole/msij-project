```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

# Проверяем подключение
print("Каталог:", spark.conf.get("spark.sql.catalog.ice.uri"))

# === 1. Подготовим DataFrame ===
data = [
    (1, "apple", 10.5, True, "2025-11-07"),
    (2, "banana", 5.2, False, "2025-11-06"),
    (3, "cherry", 8.9, True, "2025-11-05"),
    (4, "date", 11.3, False, "2025-11-04"),
    (5, "elderberry", 7.7, True, "2025-11-03"),
]

columns = ["id", "name", "price", "in_stock", "date"]

df = spark.createDataFrame(data, columns)

# === 2. Убедимся, что namespace существует ===
spark.sql("CREATE NAMESPACE IF NOT EXISTS ice.test")

# === 3. Пишем таблицу в Iceberg через REST ===
table = "ice.test.random_table"

df.writeTo(table).using("iceberg").createOrReplace()

print("✅ Таблица успешно записана:", table)

# === 4. Читаем обратно ===
df2 = spark.read.table(table)
df2.show()
```
